## Linking the GUI to a Spectrometer
The GUI is an object that links to a Spectrometer object from seabreezy. Since we are simulating the data, we will pretend to link by using a "spectrometer" which is actually just a array of wavelengths that will be used to simulate the data. Starting the GUI is a simple 3 lines:

```python
fake_spectrometer = np.linspace(200,800,1024)
gui = PL_GUI(fake_spectrometer) #create the GUI object
gui.show() #call the object's method to open the GUI itself
```

## Using the GUI
Hopefully above you, there is now a GUI! The GUI starts out by default in a static mode. Once you click "Live," it will begin collecting data of a fake "ligand exchange" happening in real time using the settings you specific in the "Exposure" and "Averages" section. This ligand exchange will increase the PL quickly and then taper off to a maximum value when equilibrium is achieved. You can also use the GUI statically to save single spectrums using the "Save Spectrum" button, which will also pin that saved spectrum to the plot for future reference. However, the live data feed also automatically saves all the spectrums recorded, so you do not ever have to use the "Save Spectrum" button.

To start, use the default settings and run the live feed (by clicking on the "Live" button), watching the Y axis, until you see the PL spectrum cap out. Then, click the "Live" button again to stop the feed and save the data.

## Data Processing
Great! We now have a basic dataset to process. The GUI automatically saves the live feed as a .csv file with the important parameters in the title, and the full spectrums plus times. We can now resolve the PL increase with time by integrating this feed using the data processing code (under construction in this file for now)

```python
data = np.loadtxt('live_data_1_200.0ms_5averages.csv', delimiter=',')

def IntegratePL(data, lowerX, upperX):
    '''Returns time-resolved PL from a live data file produced by the PL GUI.'''
    lowerI = np.abs(data[:,0]-lowerX).argmin()
    upperI = np.abs(data[:,0]-upperX).argmin()
    integration_array = data[lowerI:upperI, 1:]
    return np.trapezoid(integration_array, axis=0)

integral_array = IntegratePL(data, 300,700)

fig, ax = plt.subplots(figsize=(6,4)) #plot output
ax.plot(data[0,1:], integral_array, label='integrated PL intensity')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Intensity (a.u)')
ax.set_yticks([])
ax.legend()
```
